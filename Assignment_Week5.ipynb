a) Create a Colab loading the numpy, pandas, seaborn, and matplotlib.pyplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

b) Load the dataset Titanic from github and see the head of the dataframe.

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/pena-rodrigo/artificial/main/train.csv')
titanic.head()
```

 c) Plot a distribution of the age.

```python
sns.distplot(titanic["Age"],kde=False)
```

d) Plot a kernel density estimation of the Fare.

```python
sns.kdeplot(titanic['Fare'], shade=True)
```

 e) Plot a joint plot of age vs Fare.


```python
sns.jointplot(x='Fare',y='Age',data=titanic)
```

 f) Plot a boxplot of survived vs age.

```python
sns.boxplot(x='Survived',y='Age',data=titanic)
```

g) Plot the same but with a swarmplot.

```python
sns.swarmplot(x='Survived',y='Age',data=titanic)
```

h) Plot a countplot of Survived

```python
sns.countplot(x='Survived',data=titanic)
```

 i) Plot two distplots with the facetgrid option where the column will be the survived and the distributions show the Age.

```python
g = sns.FacetGrid(data=titanic, col='Survived')
g.map(sns.distplot, 'Age',kde=False)
```

j) Plot a heatmap using the correlation matrix, use the color "flare" for the cmap

```python
tc = titanic.corr()
sns.heatmap(tc,cmap='flare')
```
